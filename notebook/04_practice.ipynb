# 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

```python
# tips데이터 불러오기

tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

```python
# 히스토그램 그리기

# 1. 기본 틀 + 1X1인 1개 그래프 그릴 격자 구성
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)

# 2. 히스토그램 만들기
axes1.hist(tips['total_bill'], bins=10)  # bins=10 -> x축 간격: 10
                                         # hist는 일변량 그래프이므로 변수한개!
# 3. 제목 설정
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

```python
# 산점도 그리기

# 1. 기본 틀 + 1X1인 1개 그래프 그릴 격자 구성
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

# 2. 산점도 만들기
axes1.scatter(tips['total_bill'], tips['tip'])  # 산점도(scatter_plot)는 이변량 그래프이므로 변수 두개!

# 3. 제목 설정
axes1.set_title('Scatter_Plot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# 박스 그래프 그리기

# 1. 기본 틀 + 1X1인 1개 그래프 그릴 격자 구성
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

# 2. 박스그래프 만들기
# boxplot은 이산형변수(명확한 구분-> ex.Female, Male)와 연속형변수(ex. 사람들의 키)를 함께 사용
axes1.boxplot([tips[tips['sex'] == 'Female']['tip'], tips[tips['sex'] == 'Male']['tip']], labels=['Female', 'Male'])
                                   # 이산형 변수 구분                            # labels로 구분! (순서맞춰야하나?)

# 3. 제목 설정
axes1.set_title('Boxplot of Tips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
```

```python
# 다변량 데이터로 다변량 그래프 그리기 - 산점도

# 1. Female, Male과 같은 문자열은 산점도 그래프에서 정수(0, 1)로 사용
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
    
# 2. recode_sex메서드가 반환한 값을 데이터 프레임에 추가!
tips['sex_color'] = tips['sex'].apply(recode_sex)

# 3. 기본 틀 + 1X1인 1개 그래프 그릴 격자 구성
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

# 4. 산점도 만들기
axes1.scatter(x=tips['total_bill'], y=tips['tip'], s=tips['size']*10, c=tips['sex_color'], alpha=0.5)
              # x축                   y축       s=size(1~10까지)  c=color(Female, Male 구분) alpha=w점의투명도

# 5. 제목 설정
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

# 단변량 그래프 그리기 - 히스토그램

```python
# 단변량 그래프 그리기 - 히스토그램

# 1. tips데이터 집합 불러오기
import seaborn as sns
tips = sns.load_dataset('tips')

#2. 히스토그램 그리기
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density plot')
```

```python
# 옵션1. 히스토그램만 나타내기(밀집도 그래프 제외 방법)
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram')
```

```python
# 옵션2. 밀집도 그래프만 나타내기
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density plot')
```

```python
#옵션3. rug그래프 (그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집정도를 표현한 그래프)
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug plot')
```

```python
# count그래프(이산값을 나타낸 그래프)
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of Days')
```

# 이변량 그래프 그리기 - 히스토그램

```python
# 이변량 그래프 그리기

# 1. 산점도 + 회귀선 그리기
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
# 단변량 그래프 그리기 - 히스토그램
```

```python
# 이변량 그래프 그리기

# 2. 산점도 그리기
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)  # fit_reg=False: 회귀선 제거
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 이변량 그래프 그리기

# 3. 산점도 + 히스토그램 그리기
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
        # fig가 있어야 전체그래프에 대한 제목이다.
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
```

```python
# 이변량 그래프 그리기

# 4. 육각그래프 그리기 (산점도 그래프의 데이터를 구분하기 쉽게 그리기)
# 특정 데이터의 갯수가 많아지면 점점 더 진한색으로 표현됨
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')  # kind='hex'가 있어야 육각그래프임!
joint.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
        # fig가 있어야 전체그래프에 대한 제목이다.
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
```

```python
# 이변량 그래프 그리기

# 5. 이차원 밀집도 그리기
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'], data2=tips['tip'], shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 이변량 그래프 그리기

# 6. 바 그래프 그리기
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total Bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
# 이변량 그래프 그리기

# 6. 바 그래프 그리기
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of average total Bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total bill')
```

```python
# 이변량 그래프 그리기

# 7. 바이올린 그래프 그리기
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Violinplot of  total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total bill')
```

```python
# 이변량 그래프 그리기

# 8. 관계 그래프 그리기
fig = sns.pairplot(tips)

# 관계그래프는 중복된 정보가 표현됨. -> 중복된 그래프가 그려지는 위치를 지정해, 원하는 그래프로 교체

# 중복 없이 다른 그래프 그리기
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)  # 산점도 그래프
pair_grid = pair_grid.map_lower(sns.kdeplot)  # 이차원 밀집도
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)  # 히스토그램과 밀집도
plt.show()
```

```python
# 4. lmplot 메서드로 4개의 데이터 그룹에 대한 각각의 그래프 그리기
anscombe_plot = sns.lmplot(x='x', y='y', data=anscombe, fit_reg=False, col='dataset', col_wrap=4)
    # col='dataset': 4분할 하는 기준  col_wrap: 그래프를 그릴 열의 최댓값(2이면 2X2, 4이면 가로로 그래프4개)
```

```python
# 그래프에 스타일 적용하기

# 1. 바이올린 그래프 그리기 - 배경 변경
sns.set_style('whitegrid')

fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
# 2. for문을 이용해 모든 스타일을 하나씩 적용한 그래프
fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
